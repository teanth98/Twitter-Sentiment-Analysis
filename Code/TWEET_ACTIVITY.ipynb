```python
columns = ["tweetcreatedts","extractedts","userid","tweetid","text","language"]
for i in arr[1:]:
    loc = "ukraine-russian-crisis-twitter-dataset-1-2-m-rows/"+i
    df = pd.read_csv(loc, compression='gzip', index_col=0,encoding='utf-8')
    df=df[columns]
    df['formatted_dt'] = df['tweetcreatedts'].apply(lambda x: x.split()[0])
    df_language_freq=df.groupby(['language','formatted_dt'],as_index=False).size().sort_values(by='size',ascending=False)
    df_distinct_users = df.groupby(['language','formatted_dt'],as_index=False).userid.nunique()
    df_combined = df_language_freq.merge(df_distinct_users, on=['language','formatted_dt'], how='left')
    df_combined = df_combined[['language', 'formatted_dt', 'size', 'userid']]
    df_combined.columns = ['language', 'formatted_dt', 'num_tweets', 'distinct_users']
    df_combined.to_csv('aggregated_tweets_new.csv',header=False,mode='a')

    
```

```python
#### ANALYTICS METRICS CREATED HERE ####
columns_new =  ['language', 'formatted_dt', 'num_tweets', 'distinct_users']
df_new_combined = pd.read_csv('aggregated_tweets_new.csv',names=columns_new).reset_index()
df_new_combined=df_new_combined[['language', 'formatted_dt', 'num_tweets', 'distinct_users']]

```

```python
df_new_combined_tw = df_new_combined.groupby(['language','formatted_dt']).num_tweets.sum().reset_index().sort_values(by='num_tweets',ascending=False)
df_new_combined_usr = df_new_combined.groupby(['language','formatted_dt']).distinct_users.sum().reset_index().sort_values(by='distinct_users',ascending=False)
```

```python
df_new_combined_final = df_new_combined_tw.merge(df_new_combined_usr,on=['language','formatted_dt'], how='left')
```
